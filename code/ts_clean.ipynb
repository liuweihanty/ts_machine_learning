```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ts_master_list = pd.read_csv("../data/Chr7_master_list.csv")
ts_master_list.head(3)
```

```python
ts_score = pd.read_csv("../data/Rank_variables_master_list.csv")
ts_score.head(3)
```

Define a function `percent_nan` to calculate the percentage of NaN values in the given list/array.

```python
def percent_nan(list):
    return (sum(np.isnan(list))/len(list))
```

```python
nan_check  = ts_score.drop(['Gene', 'Entrez.GeneID', 'ENSEMBL_ID'], 
                           axis=1).apply(percent_nan)
nan_check.sort_values(ascending = False).head(20)
```

Drop the 5 columns that have 75% or more NaN values

```python
ts_score_cleaned_nan = ts_score.drop(nan_check[nan_check > 0.75].index.tolist(), axis=1)
```

Calculate the variance of each col.

```python
col_var = ts_score_cleaned_nan.drop(['Gene', 'Entrez.GeneID', 'ENSEMBL_ID'], 
                           axis=1).apply(np.var)
col_var.sort_values(ascending = False)
```

```python
ts_score_cleaned_var = ts_score_cleaned_nan.drop(col_var[col_var < 10e-3].index.tolist(), axis=1)
ts_score_cleaned_var.head()
```

```python
def percent_zero(list):
    return ((len(list) - np.count_nonzero(list))/len(list))
```

```python
zeros_check = ts_score_cleaned_var.drop(['Gene', 'Entrez.GeneID', 'ENSEMBL_ID'], axis=1).apply(percent_zero)
zeros_check.sort_values(ascending = False).head(20)
```

```python
ts_score_cleaned_zero = ts_score_cleaned_var.drop(zeros_check[zeros_check > 0.5].index.tolist(), axis=1)
ts_score_cleaned_zero = ts_score_cleaned_zero.drop(['Sabbatini2017.MV411.threshold'], axis=1)
```

```python
col_var_clean = ts_score_cleaned_zero.drop(['Gene', 'Entrez.GeneID', 'ENSEMBL_ID'], 
                           axis=1).apply(np.var)
col_var_clean.sort_values(ascending = False)
```

```python
ts_score_cleaned_zero.to_csv(r'../data/Rank_variables_master_list_cleaned.csv')
```
